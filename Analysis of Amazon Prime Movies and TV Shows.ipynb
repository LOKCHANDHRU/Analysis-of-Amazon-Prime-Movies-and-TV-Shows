# Step 2: Data Cleaning

```python
# Gets General information about the data before handling:

df.info()
```

- It indicates that there are total of 9668 movie details among which 2083 movies Director is missing, 1233 movies Cast details are also missing, 8996 movies Country details are also missing, 9513 movies date_added details were also missing and 337 movies rating details are also missing.

```python
# Checking for null values before handling:

df.isnull().sum()
```

```python
# Heatmap before handling:

sns.heatmap(df.isnull())
```

```python
# Checking the duplicate records:

df.duplicated().sum()
```

```python
# Creating the Data Dictionary with first column being datatype:

Data_dict = pd.DataFrame(df.dtypes)
Data_dict
```

```python
# identifying the missing values from the dataset:

Data_dict['MissingVal'] = df.isnull().sum()
Data_dict
```

```python
# Dropping the unneccesary columns:

df.drop(['show_id'], axis = 1, inplace = True)
df.drop(['date_added'], axis = 1, inplace = True)
df.drop(['description'], axis = 1, inplace = True)
```

```python
# Handling missing value:

df['director'].fillna('Not mention', inplace=True)
df['cast'].fillna('Not Mention', inplace=True)
df['country'].fillna('Not Mention', inplace=True)
df['rating']=df['rating'].fillna(df['rating'].mode()[0])
```

```python
# Checking missing values after handling:

df.isnull().sum()
```

# Step 3: Univariate Analysis

## 3.1 Count of shows by type (movie vs tv show) 

```python
# Count of shows by type:

type_counts = df['type'].value_counts()
print("Count of shows by type:")
print(type_counts)
```

```python
# in pie chart for count of shows by type (movie vs tv show):

plt.figure(figsize=(8, 6))
type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Content Types')
plt.ylabel('')
plt.show()
```

## 3.2 Most common genres or categories

```python
# Top genres:

top_genres = df['listed_in'].value_counts().head(10)
top_genres
```

```python
# in bar chart for top genres or categories:

plt.figure(figsize=(12, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='muted')
plt.title('Top 10 Genres')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

## 3.3 Frequency of ratings

```python
# Counts of Rating: 

rating_counts = df['rating'].value_counts()
print("\nCounts of rating:")
print(rating_counts)
```

```python
# in bar chart for Frequency Ratings:

rating_counts.plot(kind='bar')
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Frequency Ratings")
plt.show()
```

# Step 4: Date & Time Insights

## 4.1 Analyze the release year distribution

```python
# release year based distribution:

if 'release_year' in df.columns:
    release_counts = df['release_year'].value_counts().sort_index()
    print("\nContent release by year:")
    print(release_counts.tail(10))  
```

```python
# in line chart for release year based distribution:

release_counts.plot(kind='line')

# line chart

plt.plot(release_counts, marker = 'o', linestyle = '-', color='b',label = 'release year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.title('Release Year Based Distribution')
plt.grid()
plt.show()
```

```python
# in bar chart trends overtime in content release:

recent_years = release_counts.tail(10)
recent_years.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.show()
```
